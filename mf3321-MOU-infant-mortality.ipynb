```python
# Categorizing states by whether their IMR is above or below the 50-state median
def categorize_by_IMR(row):
    '''(number) -> string
    Return whether a State's infant mortality rate is above or below the US mean.
    '''
    if row['Infant_Mortality_Rate'] > IMR_mean:
        return 'Above'
    elif row['Infant_Mortality_Rate'] < IMR_mean :
        return 'Below'

df['IMRComparata'] = df.apply(categorize_by_IMR, axis=1)

IMR_data = df[['State', 'Infant_Mortality_Rate', 'IMRComparata']]
IMR_data.head()
```

```python
# Finding the CI for proportion for above-mean IMR and above-mean child poverty rate
def CI_prop(s, n):
    p = s / n
    se2 = p * (1-p) / n
    se = np.sqrt(se2)
    low = p - 1.96 * se
    up = p + 1.96 * se
    print('Sample proportion is:', p)
    print('95% CI is: ({}, {})'.format(low, up))
    
CI_prop(16,23)
```

```python
# Finding the CI for proportion for above-mean IMR and below-mean child poverty rate
def CI_prop(s, n):
    p = s / n
    se2 = p * (1-p) / n
    se = np.sqrt(se2)
    low = p - 1.96 * se
    up = p + 1.96 * se
    print('Sample proportion is:', p)
    print('95% CI is: ({}, {})'.format(low, up))
    
CI_prop(10,27)
```

```python
# Performing chi-square test for IMR comparata and child poverty comparata
chi2, p, dof, expected = stats.chi2_contingency(table)
print("chi2:", chi2)
print("p:", p)
print("dof:", dof)
print("expected:", expected)
```

```python
# Calculating 95% CI for proportion difference for IMR and child poverty
def CI_prop_diff(s1, s2, n1, n2):
    p1 = s1 / n1
    p2 = s2 / n2
    se2p1 = p1 * (1 - p1) / n1
    se2p2 = p2 * (1 - p2) / n2
    se2 = se2p1 + se2p2
    se = np.sqrt(se2)
    low = (p1 - p2) - 1.96 * se
    up = (p1 - p2) + 1.96 * se
    print("Proportion difference is:", p1 - p2)
    print("95% CI is: ({}, {})".format(low, up))
    
CI_prop_diff(s1=10, s2=16, n1=27, n2=23)
```

```python
# Categorizing states by 1860 status
def categorize_by_1860_status(row):
    '''(number) -> string
    Return whether a State should be counted as a slave state or a free state
    for the purposes of the chi-square test
    '''
    if row['1860_Status'] == "Free":
        return 'Free'
    if row['1860_Status'] == "NotApplicable":
        return 'Free'
    if row['1860_Status'] == "OpenSlave":
        return 'Slave'
    elif row['1860_Status'] == "Slave":
        return 'Slave'
    
df['1860_status_2_cat'] = df.apply(categorize_by_1860_status, axis=1)

IMR_data = df[['State', '1860_Status', '1860_status_2_cat']]
IMR_data.head()
```

```python
# Performing means test for slave states
def CI_mean(data):
    se = np.std(data)/np.sqrt(len(data))
    low = np.mean(data)-1.96*se
    up = np.mean(data)+1.96*se
    print("Average is:", np.mean(data))
    print("95% CI of mean is: ({}, {})".format(low,up))

CI_mean(slave_states['Infant_Mortality_Rate'])
```

```python
# Performing one sample t-test for slave states
t_val, p_val = stats.ttest_1samp(slave_states['Infant_Mortality_Rate'].dropna(), 6.14)
print("Test statistic:", t_val)
print("p-value:", p_val)
```

```python
# Performing means test for free states
CI_mean(free_states['Infant_Mortality_Rate'])
```

```python
# Performing one sample t-test for free states
t_val, p_val = stats.ttest_1samp(free_states['Infant_Mortality_Rate'].dropna(), 6.14)
print("Test statistic:", t_val)
print("p-value:", p_val)
```

```python
# Performing means difference test for slave and free states
def CI_mean_diff(list1, list2):
    s1 = np.var(list1)
    s2 = np.var(list2)
    n1 = len(list1)
    n2 = len(list2)
    se2 = s1/n1 + s2/n2
    se = np.sqrt(se2)

    diff = np.mean(list1) - np.mean(list2)
    low = diff - 1.96 * se
    up = diff + 1.96 * se
    print("The average difference is:", diff)
    print("The 95% CI is: ({}, {})".format(low, up))
    
CI_mean_diff(slave_states['Infant_Mortality_Rate'], free_states['Infant_Mortality_Rate'])
```

```python
# Performing two sample t-test
t_val, p_val = stats.ttest_ind(slave_states['Infant_Mortality_Rate'], free_states['Infant_Mortality_Rate'], equal_var=False)

print("Test statistic:", t_val)
print("p-value:", p_val)
```
